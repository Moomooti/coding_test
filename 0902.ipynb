```python
#과제 1 : 52장의 카드에서 만들 수 있는 페어가 정확히 하나만 있는 5장 조합을 모두 출력하는
#프로그램을 작성하라. 출력이 너무 많으면 줄이셈.
```

```python
import itertools
from collections import Counter

# 카드 정의 (A,2,3,...,10,J,Q,K) x (♠,◆,♥,♣)
suits = ['S', 'D', 'H', 'C']
# 스페이드, 다아이몬드, 하트, 클로버를 리스트로 만들기
ranks = ['A'] + [str(i) for i in range(2, 11)] + ['J', 'Q', 'K']
#A와 JQK 그리고 2~10까지의 숫자 리스트 만들기
deck = [rank + suit for rank in ranks for suit in suits]
#위의 2개의 조합을 사용하여, AS : 1+스페이드 등의 카드 52장 만들기

pair_comb = []

# 모든 5장 조합
# 52장에서 5장을 중복 없이 뽑는 경우를 만든다.
for hand in itertools.combinations(deck, 5):
    #슬라이싱은 카드 문자열에서 마지막 한 글자를 제외하고 앞부분만 추출되기 때문이다.
    values = [card[:-1] for card in hand]
    counter = Counter(values)
    counts = sorted(counter.values(), reverse=True)
    # 페어 1개(2,1,1,1)
    if counts == [2,1,1,1]:
        pair_comb.append(hand)

# 결과 출력 (처음 10개만 출력)
for h in pair_comb[:10]:
    print(h)
print(f"총 조합 개수: {len(pair_comb)}")

# 결과는 너무 많으니 8장 초과 시 제한해서도 구현
# 8장으로 진행
short_deck = [rank + suit for rank in ranks[:2] for suit in suits]  # A,2만 사용, 8장
short_pair_comb = []
for hand in itertools.combinations(short_deck, 5):
    values = [card[:-1] for card in hand]
    counter = Counter(values)
    counts = sorted(counter.values(), reverse=True)
    if counts == [2,1,1,1]:
        short_pair_comb.append(hand)

# 8장으로 만든 첫 10개 출력
for h in short_pair_comb:
    print(h)
print(f"(8장 사용) 총 조합 개수: {len(short_pair_comb)}")
```

```python
count = 0
for x in range(1, 100):
    for y in range(1, 100):
        z = 100 - x - y
        if z >= 1:
            print(x, y, z)
            count +=1
print("총 해의 개수:", count)
```

```python
#과제 3
# m개의 원소를 가진 집합에서 n개의 원소를 가진 집합으로 가는 전사함수의 개수를 출력하는 프로그램을 작성하라.
# m과 n의 값을 바꾸어 보면서 값이 너무 커지지 않는 입력의 범위가 어느정도인지 확인해 보라.
```

```python
import math
#m개에서 n개로 가는 전사 함수의 개수를 반환
#포함-배제 원리를 적용
def surjective_count(m, n):
    res = 0
    for k in range(n+1):
        res += ((-1)**k) * math.comb(n, k) * ((n-k)**m)
        #comb -> n개의 원소 중 k개를 고르는 경우의 수
        #n-k개의 값으로 m개의 원소를 맵핑할 수 있는 전체 함수의 개수
    return res

# m, n 바꿔가면서 출력
for m in range(1, 10):
    for n in range(1, 10):
        cnt = surjective_count(m, n)
        if cnt > 0 and cnt < 10**6:
            print(f"m={m}, n={n}, 전사함수 개수: {cnt}")
```

```python

#과제 4
#피보나치 수열을 계산하는 3가지 방법을 모두 작성해보고 실행시간을 비교하라.
#결과 값이 빨리 커지는 것을 주의할것
```

```python
import time

def fib_recursive(n):
    if n <= 2:
        return 1
    return fib_recursive(n-1) + fib_recursive(n-2)

def fib_memo(n, memo={}):
    if n <= 2:
        return 1
    if n in memo:
        return memo[n]
    memo[n] = fib_memo(n-1, memo) + fib_memo(n-2, memo)
    return memo[n]

def fib_iter(n):
    a, b = 1, 1
    for _ in range(3, n+1):
        a, b = b, a+b
    return b

N = 30

# 순수 재귀
start = time.time()
res1 = fib_recursive(N)
t1 = time.time() - start

print(f"[순수 재귀] 값: {res1}, 실행시간: {t1:.6f} 초")

# 메모이제이션 재귀
start = time.time()
res2 = fib_memo(N, {})
t2 = time.time() - start

print(f"[메모이제이션] 값: {res2}, 실행시간: {t2:.6f} 초")

# 반복문
start = time.time()
res3 = fib_iter(N)
t3 = time.time() - start

print(f"[반복문]     값: {res3}, 실행시간: {t3:.6f} 초")
```
